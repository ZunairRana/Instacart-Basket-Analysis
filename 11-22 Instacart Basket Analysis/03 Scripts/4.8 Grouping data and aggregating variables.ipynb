# Q4. Create a loyalty flag for existing customers using the transform() and loc() functions

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge[['user_id', 'max_order', 'order_number']].head()
```

```python
# Deriving 'loyalt_flag' column using loc()
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts()
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(30)
```

# Q5. Does the price of products purchased by loyal customers differ from regular or new customers


```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean', 'min', 'max', 'sum']})
```

# Q6. 

```python
#Group data based on user_id, and transform prices to generate average_price column
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(30)
```

```python
#Create flag that assigns a “types of spenders” label to a user ID based on average prices.

ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spender_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spender_flag'] = 'High spender'
```

```python
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

# Q7 In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers.

```python
#Group data based on user_id, and transform days_since_prior_order to generate median_days_since_prior_order column
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Derive column conditions
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order']>20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order']>10) & (ords_prods_merge['median_days_since_prior_order']<=20), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order']<=10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'spender_flag', 'frequency_flag']].head(30)
```

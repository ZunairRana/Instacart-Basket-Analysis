#### There are 11259 missing values for the first_name column (~5% of the data set). Although we could potentially remove the missing data without having too much of an impact on the outcome of our analysis, I will leave the missing values as we don't need the first_name to identify customers as we have the user_id column to identify them instead.

```python
#Checking for columns with mixed data types
for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

#### First_name column has a mixed data type

```python
#Check the data type of 'first_name'
df_cust['first_name'].dtype
```

```python
#Convert first_name data type to string
df_cust['first_name'] = df_cust['first_name'].astype('str')
```

```python
#Re-run check for columns with mixed data types
for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

# 04 Reduce data types to save RAM 

```python
df_cust['age'] = df_cust['age'].astype('int8')
```

```python
df_cust['no_of_dependants'] = df_cust['no_of_dependants'].astype('int8')
```

```python
df_cust['income'] = df_cust['income'].astype('int32')
```

```python
df_cust['user_id'] = df_cust['user_id'].astype('str')
```

```python
#Check for updated data types
df_cust.dtypes
```

# 05 Merge df_cust dataframe with df_ords_prods_merge

#### Common column between df_cust and df_ords_prods_merge is user_id

```python
#Check user_id data type in df_ords_prods_merge
df_ords_prods_merge['user_id'].dtype
```

```python
#Convert user_id in df_ords_prods_merge to string
df_ords_prods_merge['user_id'] = df_ords_prods_merge['user_id'].astype('str')
```

```python
df_ords_prods_merge.dtypes
```

```python
#Merge df_ords_prods_merge and df_cust
df_ords_prods_cust = df_ords_prods_merge.merge(df_cust, on = 'user_id')
```
